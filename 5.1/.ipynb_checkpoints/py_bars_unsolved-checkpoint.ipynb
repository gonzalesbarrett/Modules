```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]

# np.arange: 5 values in bars_in_cities so creating an array 0-4
x_axis = np.arange(len(bars_in_cities))
x_axis
```

```python
# Create a bar chart based upon the above data
plt.bar(x_axis, bars_in_cities, color = "b", align = "center")
```

```python
# Create the ticks for our bar chart's x axis
# Very fancy way of getting a 0 - 4 array (from example not mine)
tick_locations = [value for value in x_axis]

# xticks(ticks, label)
plt.xticks(tick_locations, cities)
```

```python
# Set the limits of the x axis xlim: (how far left, how far right (sort of)) change -.75 to 0 for example
plt.xlim(-.75, len(x_axis)-.25)
```

```python
# Set the limits of the y axis (from example) ylim(left, right)
plt.ylim(0, max(bars_in_cities)+.4)
```

```python
# Give the chart a title, x label, and y label
plt.xlabel("Cities")
plt.ylabel("Bars Per 10,000 Households")
plt.title("Density of Bars in Cities")
```
