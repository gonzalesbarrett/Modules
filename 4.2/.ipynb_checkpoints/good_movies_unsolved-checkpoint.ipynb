```python
# Dependencie
import pandas as pd
```

```python
# Load in file
movie_file = "../Resources/movie_scores.csv"
```

```python
# Read and display the CSV with Pandas
movie_data_df = pd.read_csv(movie_file)
movie_data_df
```

```python
# List all the columns in the table
column_list = list(movie_data_df)
column_list
```

```python
# We only want IMDb data, so create a new table that takes the Film and all the columns related to IMDb
imdb_columns = ['FILM','IMDB', 'IMDB_norm', 'IMDB_norm_round','IMDB_user_vote_count']
movie_data_df.loc[:, imdb_columns]
```

```python
# We only like good movies, so find those that scored over 7, and ignore the norm rating
good_movies = movie_data_df.loc[movie_data_df ['IMDB'] > 7, ["FILM", "IMDB", "IMDB_user_vote_count"]]
good_movies.head()
```

```python
# Find less popular movies--i.e., those with fewer than 20K votes
unknown_movies = good_movies.loc[good_movies["IMDB_user_vote_count"] < 20000, ['FILM', 'IMDB', 'IMDB_user_vote_count']]
unknown_movies
```
